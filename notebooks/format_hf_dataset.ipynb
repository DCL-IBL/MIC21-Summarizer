```python
from datasets import Dataset, DatasetDict
from datasets import load_dataset

import os

import matplotlib.pyplot as plt
import matplotlib.patches as patches
from matplotlib.lines import Line2D
import matplotlib.image as mpimg

import pickle
import json
import shutil
from datasets import ClassLabel
```

```python
def process_coco_data(coco_data):
    coco_data_flat = {im["file_name"]:{"width":im["width"],"height":im["height"],"annotations":[]} for im in coco_data["images"]}
    image_id_to_file_name = {im["id"]:im["file_name"] for im in coco_data["images"]}
    cat_id_to_cat_name = {cat["id"]:cat["name"] for cat in coco_data["categories"]}
    for ann in coco_data["annotations"]:
        image_id = ann["image_id"]
        cat_id = ann["category_id"]
        fname = image_id_to_file_name[image_id]
        cname = cat_id_to_cat_name[cat_id]
        coco_data_flat[fname]["annotations"].append({"cat":cname,"bbox":ann["bbox"]})
    return coco_data_flat
```

```python
domains = os.listdir("./downloaded_images")
f_meta = open(f"./mic21_dataset/train/metadata.jsonl","w")
for domain_name in domains:
    img_path = f"./downloaded_images/{domain_name}/data"
    shutil.copytree(img_path,f"./mic21_dataset/train/{domain_name}",dirs_exist_ok=True)
    
    entries = os.listdir(img_path)
    
    desc_path =  f"./descriptions/{domain_name}.pkl"
    with open(desc_path, 'rb') as f:
        desc_data = pickle.load(f)

    coco_path = f"./downloaded_json/{domain_name}_gt.json"
    with open(coco_path,"r") as f:
        coco_data = json.loads(f.read())
    coco_data_flat = process_coco_data(coco_data)
    
    for entry in entries:
        if entry in desc_data.keys():
            try:
                bufj = json.loads(desc_data[entry])
                title = bufj["Title"]
                desc = bufj["Description"]
                width = coco_data_flat[entry]["width"]
                height = coco_data_flat[entry]["height"]
                wh = [width,height]
                anns = coco_data_flat[entry]["annotations"]
                dat_out = {"file_name":domain_name+"/"+entry,"label": domain_name,"title":title,"description":desc,"size":wh,"annotations":anns}
                dat_out_json = json.dumps(dat_out).encode("utf-8").decode("utf-8")
                f_meta.write(dat_out_json+"\n")
            except:
                print("Skip "+entry)
        #img_name = os.path.join(dir_path, entry)
        #img = mpimg.imread(img_name)
        #image_bytes = open(image_path, "rb").read()
    print(domain_name)
f_meta.close()
```
